## Background and Prior Work

Predicting student dropout in higher education has become a prevalent topic in educational research because early identification of at-risk students can enable universities and colleges to proactively support their students and try to prevent them leaving the college. Student attrition is a large loss to higher education institutions, as it represents lost tuition revenue, reduced completion metrics that reflect poorly on the institution itself, and an inefficient allocation of resources. Arguably, it is sometimes worse off for students, who may incur financial debt, delayed career entry, and negative psychological consequences from leaving college early. At a higher societal level, dropout undermine the workforce and its development, and only further education and economic inequality. 

This understanding has motivated more and more work to help model student dropout risk using early academic data, serving as a basis for data driven intervention strategies to help mitigate these fallbacks of dropout. Previous research suggests that much of dropout experienced in the first years of college, and is related to students' academic performance early on in their education as it could affect their belief in their academic fit and future success. This is believed to be because early academic success and/or failures act as a feedback loop to update student beliefs in their own abilities. Furthermore, when students fail to meet expectations early on, their assessment of whether or not higher education is a worthwhile financial investment is also called into question, leading to a higher chance of dropout.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1)

Building on this theoretical research, researchers in educational data mining have this task of dropout prediction as a supervised machine learning problem. Studies applying traditional ML classification models such as logistic regression, decision trees, and boosting methods have found that academic performance serves as a relatively consistent predictor of dropout (as a binary classification task), amplified when the scope is purely on the first year of dropout. <a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2). This study highlights the feasibility of utilizing a more traditional ML approach to accurately map out dropout rates, and motivates us to focus on early academic indicators. 

Expanding on this, a similar paper from UCI expands this to a multi-class classification task, adding three labels: graduated, dropped out, or still continuing the degree after the expected amount of time. They found similar results, arguing that early academic performance is a consistent predictor of student outcomes, but they also note that many other factors and variable do play a role, for example one being their financial situation and socioeconomic status. <a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3). This study provides both a validated and cleaned dataset as well as a methodological reference point for modeling dropout.

## Research Question

To what extent does early academic performance predict student dropout among students in the Portuguese and Spanish higher education institutions? Specifically, using students’ first-year GPA, course completion rates, and credit accumulation as predictors, can we model the probability that a student drops out within one year within these institutional contexts?

## Hypothesis


We hypothesize that there will be a correlation between early academic performance of a student during their first year of university and the likelihood of dropping out. Specifically, we define poor early academic performance using measurable indicators including lower first-year GPA, lower pass rate, and fewer total credits passed. For example, we can classify lower GPAs and pass rates as those in the bottom quartile of our data. Students who meet one or more of these criteria are expected to have a higher probability of dropping out compared to students who meet standard academic progress benchmarks. However, we anticipate that this relationship will be moderate rather than strong, as academic performance alone is unlikely to fully explain dropout. Other factors such as socioeconomic background, nationality, scholarship status, and more may have an equal or greater influence, but these varaibles are out of the scope of our analysis.

The variables from this table range from normal metrics such as age or gender to more specific ones such as admission grades or frequency of attendance. For our project, we'll mostly just be looking at the variables related to academic performance, but there are a few that might provide interesting information that we'll also keep an eye on. Early academic performance could also be influenced by other factors, such as previous qualifications or admission grades found in the dataset. The dataset contains information on the number of units a student takes in their first year, split into two semesters. The grade averages are the GPA of a specific student in that specific semester, which just measures the academic performance of a student. The GPA is on the Portuguese scale, so their equivalent of a 4.0 GPA would be a 20.0. A bad gpa would be considered anything less than a 14.0 or the equivalent of a 2.0 in the U.S. A proficient GPA would be between 14.0 and 16.0, which is below a 3.0 GPA. Anything above a 16.0 would just be considered a good GPA. The number of units enrolled is just a numerical value that measures how many credit units/hours a student is taking. This data can be compared to the column with the approved credit units, since that column measures the credits a student earns from successfully passing their courses. The number of evaluations a student takes refers to the number of  exams (evaluations) a student takes in a semester. There is a  final column called "target", which pretty much lists the final status of these students after conducted for the study, categorical data that lists either "enrolled", "graduate", or "drop-out". These 6 main columns provide some information on the early academic performance of a student through GPA and curricular units, which are about 25 to 30 hours of work per unit. 
- Enrolled: Number of curricular units being taken in a semester

### Dataset 1 Exploratory Data Analysis

```python
## Convert two GPA columns to 4.0 scale for our understanding
## Also, average GPAs over both semesters and add as a new column
max_port = 20.0
min_port = 0.0
max_usa = 4.0

grade_cols = [
    'Sem1_Gpa',
    'Sem2_Gpa'
]

for col in grade_cols:
    data1[col] = (data1[col] / max_port) * max_usa

data1['Avg_Gpa'] = (data1['Sem1_Gpa'] + data1['Sem2_Gpa']) / 2

print(data1[['Sem1_Gpa','Sem2_Gpa', 'Avg_Gpa']].head(5))
```

```python
## Totaling up enrolled and passed credits for both semesters, including a ratio between the two

data1['Total_Enrolled'] = data1['Sem1_Enrolled'] + data1['Sem2_Enrolled']
data1['Total_Passed'] = data1['Sem1_Passed'] + data1['Sem2_Passed']
data1['Pass_Rate'] = data1['Total_Passed']/data1['Total_Enrolled']

print(data1[['Total_Enrolled','Total_Passed', 'Pass_Rate']].head(5))
data1
```

#### Risk Stratification/Categorization Plots

```python
# Construct Dropout Binary Target
data1['Dropout_Binary'] = (data1['Target'] == 'Dropout').astype(int)

# GPA Quartiles
data1['GPA_Quartile'] = pd.qcut(data1['Avg_Gpa'], 4)

gpa_dropout_1 = data1.groupby('GPA_Quartile', observed=False)['Dropout_Binary'].mean()

gpa_dropout_1.plot(kind='bar')
plt.ylabel("Dropout Rate")
plt.xlabel("Average GPA Quartile Ranges (4.0 Scale)")
plt.title("Dropout Rate by GPA Quartile (Dataset 1)")
plt.show()

# Pass Rate Groups (manual bins)
bins = [0, 0.5, 0.75, 0.9, 1.0]
labels = ["0-50%", "50-75%", "75-90%", "90-100%"]

data1['PassRate_Group'] = pd.cut(
    data1['Pass_Rate'],
    bins=bins,
    labels=labels,
    include_lowest=True
)

passrate_dropout_1 = data1.groupby('PassRate_Group', observed=False)['Dropout_Binary'].mean()

passrate_dropout_1.plot(kind='bar')
plt.ylabel("Dropout Rate")
plt.xlabel("Pass Rate Bins")
plt.title("Dropout Rate by Pass Rate Group (Dataset 1)")
plt.show()

data1['TotalPassed_Quartile'] = pd.qcut(
    data1['Total_Passed'],
    4,
    duplicates='drop'
)

totalpassed_dropout_1 = data1.groupby('TotalPassed_Quartile', observed=False)['Dropout_Binary'].mean()

totalpassed_dropout_1.plot(kind='bar')
plt.ylabel("Dropout Rate")
plt.xlabel("Total Passed Quartile Ranges")
plt.title("Dropout Rate by Total Passed Quartile (Dataset 1)")
plt.show()
```

There is a decrease in dropout rate (for the most part) that can be seen in the number of credits passed and the overall pass rate. Students who have a solid foundation at the start of their academic careers are usually set for the rest of their university, with a stronger foundation resulting in a lower dropout rate. 

One odd thing in this graph is that there is a marginally higher rate of dropout for those who pass more credits in the uppermost quartile compared to the next quartile. It seems that at this point that dropout rates are indistinguishable past a certain amount of completed credits (around 10 credits at this school).

### Analysis #1: Baseline Inferential Statistics

We first perform Welch’s independent two-sample t-tests between our main predictors mentioned in the hypothesis (GPA, Credits Passed, and Pass Rate) and dropout, in order to determine whether there exists a statistically significant difference in early academic performance between students who eventually graduate and those who drop out. By establishing these significant differences, we provide a simple statistical foundation for the predictive models developed in the subsequent section.

We utilize Welch's independent t-test over the typical Student's t-test because we cannot make the assumption that the variances of student GPA, credits passed and their pass rate are the same between students who dropped out and students who didn't. Furthermore, we assume the sampling distribution of the mean is approximately normal due to the size of each dataset following the Central Limit Theorem. 


```python
def perform_t_tests(df, col_names, target_col_name):
    """
    Performs Welch's t-tests on a list of columns against a binary target.
    Displays results in a formatted pandas DataFrame.
    """
    results = []
    
    # Identify unique groups in the target column, ensure we are using dropout binary
    groups = df[target_col_name].unique()
    if len(groups) != 2:
        raise ValueError(f"Target column must have exactly 2 classes. Found: {groups}")
    
    group1 = df[df[target_col_name] == groups[0]]
    group2 = df[df[target_col_name] == groups[1]]
    
    for col in col_names:
        t_stat, p_val = stats.ttest_ind(
            group1[col], 
            group2[col], 
            equal_var=False # Welch's
        )
        
        results.append({
            'Predictor': col,
            f'Mean ({groups[0]})': group1[col].mean(),
            f'Mean ({groups[1]})': group2[col].mean(),
            'T-Statistic': t_stat,
            'P-Value': p_val
        })
    
    # Create and format the table
    results_df = pd.DataFrame(results).set_index('Predictor')
    return results_df.style.format({
        'T-Statistic': '{:.4f}',
        'P-Value': '{:.4e}'
    }).background_gradient(subset=['P-Value'], cmap='Reds_r')

```

```python
predictors = ['Avg_Gpa', 'Total_Passed', 'Pass_Rate']
results_table = perform_t_tests(data1, predictors, 'Dropout_Binary')
print("Dataset #1 Welch's t-test Results")
display(results_table)
```

```python
predictors = ['GPA', 'Total_Passed', 'Pass_Rate']
results_table = perform_t_tests(data2, predictors, 'Dropout_Binary')
print("Dataset #2 Welch's t-test Results")
display(results_table)
```

As can be seen from the tables above, both datasets have unanimous statistical significance across all three predictors at a p value of 0.001. As was seen in our EDA and now confirmed in our t-tests, non-dropout students consistently outperform dropout students by a large margin through all academic performance indicators, and have established that these predictor variables can serve as our essential features in our next section, where we build our models. 

### Analysis #2: Predictive Models

To analyze the extent for which our predictor variables can model dropout, we implement several classical machine learning models. As according to our research question, we utilize predictors related to the number of credits passed, the GPA, and the percentage of credits passed. For dataset #1, this will be a more granular set of predictors, as information such as semester GPAs are available, whereas in dataset #2, this information is only available accross an entire year. As such, we can already theorize that models trained under dataset #1 will perform slightly better compared to a model predicting on dataset #2. 

Another item to note is the hugely imbalanced class sizes. For example, in dataset #2 in particular, we have 6096 non-dropout students and 734 students who dropped out, which is a 90/10 split. As such, along with the rough sizes of our data, we chose to implement a stratified cross validation method to produce folds that do not underrepresent the dropout label. We also utilize class weights set to balanced when applicable with our models to also try to alleviate this classi imbalance.

Another point of discussion is test metrics to evaluate on. As aforementioned, because of the class imbalance, accuracy becomes less of a viable metric to verify model performance. For example, with dataset #2, models will achieve a 90% accuracy by just guessing that a student won't dropout, but this isn't ideal. As such, we heavily rely on precision, recall, and overall f1 scores, which account for this class imbalance by looking at false positives and false negatives of both labels. In the context of our research question and overall goals, the most important metric is the recall score for the dropout label (i.e. 1). This is because we would like identify at-risk students, so maximizing the model's effiency at flagging students at-risk is ideal. However, it would be best if this does not come at the cost of precision.

**ADD SOME COMMENTARY ON THE TYPES OF MODELS WE CHOSE AND WHY**

These observations allow us to come to a final, overall pipeline for model prediction. First, we apply scaling whenever necessary, as some of our models that we will utilize (such as SVMs) are succeptible to scaling issues between predictors. We will also be performing validation using stratified k-fold validation as previously mentioned. Next, we perform model and hyperparameter selection. We utilize a GridSearch to do this for us, analyzing each possibility by checking it's overall f1 score. This GridSearch will help yield the best performing model and set of hyperparameters specific to that model on the f1 score. Within our GridSearch, we also perform a simple feature selector that selects the k-best features, and this hyperparameter is also tested within our GridSearch. 



```python
def run_model(model_name, pipeline, param_grid, X_train, y_train, X_test, y_test, cv):
    """
    Runs GridSearchCV for a given model pipeline and prints evaluation results.
    """

    print("\n" + "="*75)
    print(f"MODEL: {model_name}")
    print("="*75)

    grid = GridSearchCV(
        estimator=pipeline,
        param_grid=param_grid,
        cv=cv,
        scoring="f1",
        n_jobs=-1
    )

    # Train model
    grid.fit(X_train, y_train)

    best_model = grid.best_estimator_

    # Test predictions
    y_pred = best_model.predict(X_test)

    print("\nBest Hyperparameters:")
    print(grid.best_params_)

    print("\nBest Cross-Validation F1 Score:")
    print(f"{grid.best_score_:.4f}")

    print("\nTest Set Results")
    print("-"*40)

    print("Classification Report:")
    print(classification_report(y_test, y_pred))

    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))

    print("="*75)

    return best_model, y_pred
```
