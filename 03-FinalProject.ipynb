## Abstract
Student dropout is a critical issue in higher education, affecting student well-being, institutional retention, and overall educational outcomes. In this project, we investigated whether early academic performance can predict student dropout by comparing two higher education datasets from Portugal and Spain. Our analysis focused on three primary first-year academic indicators: GPA, pass rate, and total credits passed.

To ensure the datasets were comparable, we cleaned and wrangled each dataset, standardized GPA scales, and constructed shared first-year academic performance measures. We then conducted exploratory data analysis (EDA) using visualizations like bar charts, boxplots, scatter plots, and correlation matrices to observe how academic performance differs by student outcome. We employed correlation analysis and Welch’s two-sample t-tests to determine if the differences in first-year performance between students who dropout and those who don't were statistically significant. Finally, we trained classical machine learning models to assess the predictive power of these academic indicators in identifying dropout. 

Across both datasets, the EDA and statistical analyses revealed a consistent pattern: students who dropped out generally had lower first-year GPAs, lower pass rates, and fewer credits passed than those who graduated or remained enrolled. These differences were statistically significant in both datasets, supporting our hypothesis that weaker first-year academic performance is strongly associated with a higher probability of dropping out.

In summary, our findings bring to attention that early academic performance is an insightful indicator of dropout risk across different higher education contexts. However, academic performance alone does not fully determine a student's outcome; some lower-performing students manage to graduate, while some higher-performing students do not due to external factors. This suggests that while first-year metrics are useful predictors, other financial, social, and institutional factors also play a significant role in student retention.

## Background and Prior Work

Predicting student dropout in higher education has become a prevalent topic in educational research because early identification of at-risk students can enable universities and colleges to proactively support their students and try to prevent them from leaving the college. Student attrition is a large loss to higher education institutions, as it represents lost tuition revenue, reduced completion metrics that reflect poorly on the institution itself, and an inefficient allocation of resources. Arguably, it is sometimes worse off for students, who may incur financial debt, delayed career entry, and negative psychological consequences from leaving college early. At a higher societal level, dropout undermines the workforce and its development, and only furthers education and economic inequality. 

This understanding has motivated more and more work to help model student dropout risk using early academic data, serving as a basis for data-driven intervention strategies to help mitigate these fallbacks of dropout. Previous research suggests that much of dropout is experienced in the first years of college, and is related to students' academic performance early on in their education as it could affect their belief in their academic fit and future success. This is believed to be because early academic success and/or failures act as a feedback loop to update student beliefs in their own abilities. Furthermore, when students fail to meet expectations early on, their assessment of whether or not higher education is a worthwhile financial investment is also called into question, leading to a higher chance of dropout.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1)

Building on this theoretical research, researchers in educational data mining have approached this task of dropout prediction as a supervised machine learning problem. Studies applying traditional ML classification models such as logistic regression, decision trees, and boosting methods have found that academic performance serves as a relatively consistent predictor of dropout (as a binary classification task), amplified when the scope is purely on the first year of dropout. <a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2). This study highlights the feasibility of utilizing a more traditional ML approach to accurately map out dropout rates, and motivates us to focus on early academic indicators. 

Expanding on this, research associated with the UCI Machine Learning Repository expands this to a multi-class classification task, adding three labels: graduated, dropped out, or still continuing the degree after the expected amount of time. They found similar results, arguing that early academic performance is a consistent predictor of student outcomes, but they also note that many other factors and variables do play a role, for example one being their financial situation and socioeconomic status. <a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3). This study provides both a validated and cleaned dataset as well as a methodological reference point for modeling dropout.

## Hypothesis


We hypothesize that there will be a correlation between early academic performance of a student during their first year of university and the likelihood of dropping out. Specifically, we define poor early academic performance using measurable indicators including a lower first-year GPA, a lower pass rate, and fewer total credits passed. For example, we can classify lower GPAs and pass rates as those in the bottom quartile of our data. Students who meet one or more of these criteria are expected to have a higher probability of dropping out compared to students who meet standard academic progress benchmarks. However, we anticipate that this relationship will be moderate rather than strong, as academic performance alone is unlikely to fully explain dropout. Other factors such as socioeconomic background, nationality, scholarship status, and more may have an equal or greater influence, but these variables are out of the scope of our analysis.

## Data

### Data overview

#### Dataset #1: Predict Students’ Dropout and Academic Success Dataset
The Students' Dropout and Academic Success dataset consists of 1 large CSV file containing a table of data from 4424 responses of students, and 36 different variables/questions. This dataset was created to identify students at risk of dropping out early in their academic career. The variables from this table range from normal metrics such as age or gender to more specific ones such as admission grades or frequency of attendance. For our project, we just look at the variables related to academic performance. The link to the data set can be found [here](https://archive.ics.uci.edu/dataset/697/predict+students%27+dropout+and+academic+success)

The columns from the dataset will look at will be the curricular units from the first and second semesters, which include a students' grade averages, units enrolled, units credited, units evaluated, units not evaluated, and units approved, as well as the target column. Early academic performance could also be influenced by other factors, such as previous qualifications or admission grades found in the dataset. The dataset contains information on the number of units a student takes in their first year, split into two semesters. The grade averages are the GPA of a specific student in that specific semester, which just measures the academic performance of a student. The GPA is on the Portuguese scale, so their equivalent of a 4.0 GPA would be a 20.0. A bad GPA would be considered anything less than a 14.0 or the equivalent of a 2.0 in the U.S. A proficient GPA would be between 14.0 and 16.0, which is below a 3.0 GPA. Anything above a 16.0 would just be considered a good GPA. The number of units enrolled is just a numerical value that measures how many credit units/hours a student is taking. This data can be compared to the column with the approved credit units, since that column measures the credits a student earns from successfully passing their courses. The number of evaluations a student takes refers to the number of  exams (evaluations) a student takes in a semester. The number of credited curricular units refers to transfer units from previous coursework. There is a final column called "target", which lists the final matriculation status of these students after the expected time taken for their degree program. This categorical data is listed as either "enrolled", "graduate", or "drop-out". These 6 main columns provide some information on the early academic performance of a student through GPA and curricular units, which are about 25 to 30 hours of work per unit. 
- Enrolled: Number of curricular units being taken in a semester
- Credited: Number of transfer curricular units from prior courses
- Evaluated: Number of exams taken in curricular units in a semester
- Not Evaluated: Number of curricular units without any exams/evaluations
- Approved: Number of curricular units passed
- Target: Status of student, enrolled, graduate, or dropout

Additionally, two columns we could also look at would be previous qualifications, which are integers that tell us the education level before entering university (a continuous integer scale from 0 to 200 is also in the dataset), and the other column is an admission grade, ranging from 0 to 200. These can also influence early academic performance. A poor admission grade would be below 100, a proficient one would be below 150, and anything higher would be considered a good admission grade reflective of their prior experience. While these are all very useful metrics and the dataset is mostly cleaned out and ready to be used, but there are a few concerns regarding this dataset. One concern is the fact that all the data is taken from Portuguese students, so indicators like GPA, course units, and even the overall academic system will differ from the USA standard. This might require making some values more readable, such as a Portuguese GPA conversion to the U.S. system. We'll just have to take into account possible differences in how Portuguese higher education varies from the U.S. system. A smaller concern would be the fact that while the dataset is clean, there are a lot of columns with 0 values that we will not be able to use. So while the dataset is tidy, there are plenty of rows that are incomplete. Besides that, this dataset has a lot of information that is ready to be used, and any concerns we have with it can be worked around. 

#### Dataset #2: University Student Dropout Dataset
The University Student Dropout dataset is organized as yearly CSV files named dataset_{year}.csv, with each row corresponding to a student-course enrollment for that academic year. There are 169 variables, a total of 39,364 students with 464,739 student-course observations across all years. Each file integrates data from four sources: students, programs, courses, and digital logs, and also groups variables into six thematic categories: context, admission pathways, socio-economic and demographic background, academic data, digital logs, and Wi-Fi access. Contextual attributes include anonymized identifiers for students, courses, academic programs, and campuses, as well as the academic year and group IDs, capturing where and how each student is enrolled. Like for dataset #1, we will mainly be looking at variables related to academic performance. The link to the dataset can be found at: [here](https://zenodo.org/records/17239943)

Academic data is the most detailed category, including grades, credits enrolled and earned across multiple years, semester performance, adjustments for credit recognition, internships, activities, and overall progress toward degree completion. Metrics like cumulative GPA, credits passed per semester, and credit completion rates across previous years allow for longitudinal assessment of academic success and dropout risk.  All variables are anonymized using hash codes, and numerical metrics such as grades are scaled linearly (e.g., 0–10 for grades), while credit counts are in academic credit units. This dataset is less prepared for analysis for our purposes compared to the first dataset, and much preprocessing is required. Since each entry in this dataset represents a student-course enrollment and only reports on the grade received for that singular course, student data must be first aggregated, then calculated across all the courses they take within their first year. Furthermore, this dataset has a similar column that represents dropout, but this is a binary label. To determine whether or not a student is still enrolled, students who were first years in the 2018 year csv file had their enrollment status checked in 2022, which was their expected graduation date. These processes are explained and shown more in depth in the data wrangling process. With these calculations, this dataset is able to yield similarly expressive data as the first dataset, with the following general column categories:
- Enrolled: Number of curricular units being taken in a semester
- Credited: Number of transfer curricular units from prior courses
- Credits_Passed_Exam: Number of exams taken in curricular units in a semester
- Credits_Passed_Special: Number of curricular units without any exams/evaluations
- Target: Status of student, enrolled, graduate, or dropout

While these metrics provide valuable insights, several concerns about the dataset should be noted. It is drawn from a single Spanish technological university, which limits generalizability to other fields or institutions, particularly in humanities or social sciences. Early dropouts may be underrepresented, and some variables, like parental education, employment, student dedication, may be self-reported and incomplete. Finally, identifiers are anonymized, which may reduce the precision of longitudinal tracking, and the data does not include periods affected by the COVID-19 pandemic, meaning it may not capture disruptions caused by virtual or hybrid learning environments. Despite these limitations, the dataset provides a detailed framework for studying factors influencing student retention and academic success.

To mitigate the shared concern of both datasets being drawn from singular sources, we analyze both datasets side by side, allowing us to try to capture a broader scope of students and educational backgrounds.

## Results

### Dataset 1 Exploratory Data Analysis

```python
# Convert two GPA columns to 4.0 scale for our understanding
# Also, average GPAs over both semesters and add as a new column
max_port = 20.0
min_port = 0.0
max_usa = 4.0

grade_cols = [
    'Sem1_Gpa',
    'Sem2_Gpa'
]

for col in grade_cols:
    data1[col] = (data1[col] / max_port) * max_usa

data1['Avg_Gpa'] = (data1['Sem1_Gpa'] + data1['Sem2_Gpa']) / 2

print(data1[['Sem1_Gpa','Sem2_Gpa', 'Avg_Gpa']].head(5))
```

```python
# Totaling up enrolled and passed credits for both semesters, including a ratio between the two
data1['Total_Enrolled'] = data1['Sem1_Enrolled'] + data1['Sem2_Enrolled']
data1['Total_Passed'] = data1['Sem1_Passed'] + data1['Sem2_Passed']
data1['Pass_Rate'] = data1['Total_Passed']/data1['Total_Enrolled']

print(data1[['Total_Enrolled','Total_Passed', 'Pass_Rate']].head(5))
data1
```

#### Risk Stratification/Categorization Plots

```python
# Construct Dropout Binary Target
data1['Dropout_Binary'] = (data1['Target'] == 'Dropout').astype(int)

# GPA Quartiles
data1['GPA_Quartile'] = pd.qcut(data1['Avg_Gpa'], 4)

gpa_dropout_1 = data1.groupby('GPA_Quartile', observed=False)['Dropout_Binary'].mean()

gpa_dropout_1.plot(kind='bar')
plt.ylabel("Dropout Rate")
plt.xlabel("Average GPA Quartile Ranges (4.0 Scale)")
plt.title("Dropout Rate by GPA Quartile (Dataset 1)")
plt.show()

# Pass Rate Groups (manual bins)
bins = [0, 0.5, 0.75, 0.9, 1.0]
labels = ["0-50%", "50-75%", "75-90%", "90-100%"]

data1['PassRate_Group'] = pd.cut(
    data1['Pass_Rate'],
    bins=bins,
    labels=labels,
    include_lowest=True
)

passrate_dropout_1 = data1.groupby('PassRate_Group', observed=False)['Dropout_Binary'].mean()

passrate_dropout_1.plot(kind='bar')
plt.ylabel("Dropout Rate")
plt.xlabel("Pass Rate Bins")
plt.title("Dropout Rate by Pass Rate Group (Dataset 1)")
plt.show()

data1['TotalPassed_Quartile'] = pd.qcut(
    data1['Total_Passed'],
    4,
    duplicates='drop'
)

totalpassed_dropout_1 = data1.groupby('TotalPassed_Quartile', observed=False)['Dropout_Binary'].mean()

totalpassed_dropout_1.plot(kind='bar')
plt.ylabel("Dropout Rate")
plt.xlabel("Total Passed Quartile Ranges")
plt.title("Dropout Rate by Total Passed Quartile (Dataset 1)")
plt.show()
```

There is a decrease in dropout rate (for the most part) that can be seen in the number of credits passed and the overall pass rate. Students who have a solid foundation at the start of their academic careers are usually set for the rest of their university, with a stronger foundation resulting in a lower dropout rate. 

One odd thing in this graph is that there is a marginally higher rate of dropout for those who pass more credits in the uppermost quartile compared to the next quartile. It seems that at this point that dropout rates are indistinguishable past a certain amount of completed credits (around 10 credits at this school).

### Analysis #1: Baseline Inferential Statistics

We first perform Welch’s independent two-sample t-tests between our main predictors mentioned in the hypothesis (GPA, Credits Passed, and Pass Rate) and dropout, in order to determine whether there exists a statistically significant difference in early academic performance between students who eventually graduate and those who drop out. By establishing these significant differences, we provide a simple statistical foundation for the predictive models developed in the subsequent section.

We utilize Welch's independent t-test over the typical Student's t-test because we cannot make the assumption that the variances of student GPA, credits passed and their pass rate are the same between students who dropped out and students who didn't. Furthermore, we assume the sampling distribution of the mean is approximately normal due to the size of each dataset following the Central Limit Theorem. 


```python
def perform_t_tests(df, col_names, target_col_name):
    """
    Performs Welch's t-tests on a list of columns against a binary target.
    Displays results in a formatted pandas DataFrame.
    """
    results = []
    
    # Identify unique groups in the target column, ensure we are using dropout binary
    groups = df[target_col_name].unique()
    if len(groups) != 2:
        raise ValueError(f"Target column must have exactly 2 classes. Found: {groups}")
    
    group1 = df[df[target_col_name] == groups[0]]
    group2 = df[df[target_col_name] == groups[1]]
    
    for col in col_names:
        t_stat, p_val = stats.ttest_ind(
            group1[col], 
            group2[col], 
            equal_var=False # Welch's
        )
        
        results.append({
            'Predictor': col,
            f'Mean ({groups[0]})': group1[col].mean(),
            f'Mean ({groups[1]})': group2[col].mean(),
            'T-Statistic': t_stat,
            'P-Value': p_val
        })
    
    # Create and format the table
    results_df = pd.DataFrame(results).set_index('Predictor')
    return results_df.style.format({
        'T-Statistic': '{:.4f}',
        'P-Value': '{:.4e}'
    }).background_gradient(subset=['P-Value'], cmap='Reds_r')

```

```python
predictors = ['Avg_Gpa', 'Total_Passed', 'Pass_Rate']
results_table = perform_t_tests(data1, predictors, 'Dropout_Binary')
print("Dataset #1 Welch's t-test Results")
display(results_table)
```

```python
predictors = ['GPA', 'Total_Passed', 'Pass_Rate']
results_table = perform_t_tests(data2, predictors, 'Dropout_Binary')
print("Dataset #2 Welch's t-test Results")
display(results_table)
```

As can be seen from the tables above, both datasets have unanimous statistical significance across all three predictors at a p value of 0.001. Combined with the fact that all T-statistics are abnormally high, we can conclude that these results are not from noisy data. As was seen in our EDA and now confirmed in our t-tests, students who don't dropout consistently outperform students who do by a large margin through all academic performance indicators, and have established that these predictor variables can serve as our essential features in our next section, where we build our models. 

### Analysis #2: Predictive Models

To analyze the extent to which our predictor variables can model dropout, we implement several classical machine learning models. As according to our research question, we utilize predictors related to the number of credits passed, the GPA, and the percentage of credits passed. For dataset #1, this will be a more granular set of predictors, as information such as semester GPAs are available, whereas in dataset #2, this information is only available across an entire year. Furthermore, from the joint distribution graphs from the EDA between GPA, pass rate, and dropout, we can see that dataset #2 has noisier data compared to dataset #1. As such, we can already theorize that models trained under dataset #1 will perform better compared to a model predicting on dataset #2. 

We frame this as a binary classification task, predicting whether a student would dropout or if they would not. While the EDA explores the multi-class version of our target label for dropout (1 for will dropout, 0 for will not dropout), because identifying whether or not a student would dropout is more useful to universities and students versus additionally predicting whether or not they would still be enrolled after their expected graduation date, we choose to frame this as a binary task. 

Another item to note is the hugely imbalanced class sizes. For example, in dataset #2 in particular, we have 6096 students who didn't dropout and 734 students who dropped out, which is a 90/10 split. As such, along with the rough sizes of our data, we chose to implement a stratified cross validation method to produce folds that do not underrepresent the dropout label. We also utilize class weights set to balanced when applicable with our models to also try to alleviate this class imbalance.

Another point of discussion is test metrics to evaluate on. As aforementioned, because of the class imbalance, accuracy becomes less of a viable metric to verify model performance. For example, with dataset #2, a dummy model will achieve a 90% accuracy by just guessing that a student won't dropout, but this isn't ideal. As such, we heavily rely on precision, recall, and overall F1-scores, which account for this class imbalance by looking at false positives and false negatives of both labels. Because the dummy model that just guesses the majority class would get a f1-score of 0, we will be able to properly assess whether our models have any actual predictive power. In the context of our research question and overall goals, the most important metric is the recall score for the dropout label. This is because we would like to identify at-risk students, so maximizing the model's efficiency at flagging students at-risk is ideal. However, it would be best if this does not come at the cost of precision.

We elected to look at many different types of models so that we could consider both linear and non-linear relationships and to see how  consistent the results were across different frameworks. The models we chose include: Logistic Regression, K Nearerst Neighbors(KNN), Support Vector Machines(SVM), Random Forest, and XGBoost. For example, we started with a Logistic Regression because it provides the ability to quickly find the impact of individual features. Next we moved to KNN because it is agnostic to the distributions of our data and it is usually good at finding local patterns. We then moved to more complex models such as the Random Forest model which uses trees to provide a model that is resistant to outliers and overfitting. 

These observations allow us to come to a final, overall pipeline for model prediction. First, we apply scaling whenever necessary, as some of our models that we will utilize (such as SVMs) are susceptible to scaling issues between predictors. We will also be performing validation using stratified k-fold validation as previously mentioned. Next, we perform model and hyperparameter selection. We utilize a GridSearch to do this for us, analyzing each possibility by checking its overall F1-score. This GridSearch will help yield the best performing model and set of hyperparameters specific to that model on the F1-score. Within our GridSearch, we also perform a simple feature selector when applicable that selects the k-best features, and this hyperparameter is also tested within our GridSearch. 



```python
def run_model(model_name, pipeline, param_grid, X_train, y_train, X_test, y_test, cv):
    """
    Runs GridSearchCV for a given model pipeline and prints evaluation results, feature importance, confusion matrix, and best hyperparameters.
    """

    print(f"MODEL: {model_name}")

    grid = GridSearchCV(
        estimator=pipeline,
        param_grid=param_grid,
        cv=cv,
        scoring="f1",
        n_jobs=-1
    )

    # Train model
    grid.fit(X_train, y_train)
    best_pipeline = grid.best_estimator_

    # The model is always the last step in the pipeline
    model_step = best_pipeline.steps[-1][1]
    
    selector_step = None
    # Find KSelector
    for name, step in best_pipeline.steps:
        if hasattr(step, 'get_support'):
            selector_step = step

    # Align Feature Names with the Selector
    if selector_step:
        mask = selector_step.get_support()
        feature_names = X_train.columns[mask].tolist()
    else:
        feature_names = X_train.columns.tolist()

    # Print Training/Testing Results
    y_pred = best_pipeline.predict(X_test)

    print("\nBest Hyperparameters:")
    print(grid.best_params_)

    print("\nBest Cross-Validation F1-score:")
    print(f"{grid.best_score_:.4f}")

    print("\nTest Set Results")
    print("Classification Report:")
    print(classification_report(y_test, y_pred))

    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))

    # Extract and Print Feature Importance
    print("\nTop Feature Importances/Coefficients:")
    
    try:
        importances = None
        # Check Tree-based importance (Random Forest, XGBoost)
        if hasattr(model_step, 'feature_importances_'):
            importances = model_step.feature_importances_
        # Check Linear coefficients (Logistic Regression, SVM with linear kernel)
        elif hasattr(model_step, 'coef_'):
            importances = np.abs(model_step.coef_[0])

        if importances is not None:
            feat_imp = pd.Series(importances, index=feature_names).sort_values(ascending=False)
            print(feat_imp.head(7))  # Shows top 7 most impactful features
        else:
            print("Feature importance not available for this model type or hyperparameter config")
            
    except Exception as e:
        print(f"Could not calculate feature importance: {e}")

    print("="*75)

    return best_pipeline, y_pred
```
